### Q2. (a)

No, g'(x(k-1)) = 0 does not guarantee finding the exact solution. Here's why:

In fixed-point iteration, we rewrite the equation f(x) = 0 as g(x) = x, and then iteratively compute x(k) = g(x(k-1)). For convergence, the condition |g'(x*)| < 1 must hold near the fixed point x*.

Case When g'(x(k-1)) = 0: If g'(x(k-1)) = 0 at a particular iteration, this indicates that the derivative of g(x) at x(k-1) vanishes, which can slow down convergence. While it reduces the linear error term, it does not guarantee the exact solution.

Reason: The convergence depends on the higher-order terms in the Taylor expansion of g(x) and the proximity of x(k-1) to the fixed point x*. If x(k-1) is far from x*, the higher-order terms may still prevent convergence to the exact solution.

### Q2. (b)

To show that convergence is quadratic, i will see the error at iteration k, e(k) = x(k) - x*, under the condition g'(x(k-1)) = 0.

The Taylor expansion of g(x) around x* gives me:
g(x) = x* + (1/2)g''(x*)(x - x*)^2 + higher-order terms.

Substituting x(k-1) into this i got:

x(k) = g(x(k-1)) = x* + (1/2)g''(x*)(x(k-1) - x*)^2 + higher-order terms.

Error Expression:

The error at iteration k is:
e(k) = x(k) - x* = (1/2)g''(x*)(e(k-1))^2 + higher-order terms.

Quadratic Convergence:

Quadratic convergence means that the error satisfies:
e(k) proportional to (e(k-1))^2.

This is clear from the error expression above, as e(k) depends quadratically on e(k-1).

Thus, when g'(x(k-1)) = 0, the iteration exhibits quadratic convergence.

```python

def evaluate_function(x):
    return x**4 - 10*x**3 + 35*x**2 - 50*x + 24

def compute_derivative(x):
    return 4*x**3 - 30*x**2 + 70*x - 50

def central_difference(f, x, step):
    return (f(x + step/2) - f(x - step/2)) / step

def richardson_extrapolation(func, point, diff_method, step, exponent1, exponent2):
    scale_factor = 2**exponent1
    refined_step = scale_factor * diff_method(func, point, step/2) - diff_method(func, point, step)
    return refined_step / (scale_factor - 1)

# Plotting

import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(0, 5, num=500)
y_values = [evaluate_function(val) for val in x_values]
y_derivative = [compute_derivative(val) for val in x_values]

# Plot the results
plt.plot(x_values, y_values, label='f(x)')
plt.plot(x_values, y_derivative, alpha=0.8, linestyle='--', label='f\'(x)')
plt.xlabel("x")
plt.ylabel("Function values")
plt.legend()
plt.show()
```

### Q6. (d)

```python

def gauss_diff(x):
    
    return -2 * x * np.exp(-x**2)

def gauss_diff_central(x, h):
    
    return (np.exp(-(x + h)**2) - np.exp(-(x - h)**2)) / (2 * h)


exact_derivative = gauss_diff(0.75)
central_derivative = []
step_sizes = []


for i in range(1, 12):
    h = 10**-i  
    step_sizes.append(h)
    approx_derivative = gauss_diff_central(0.75, h=h)
    central_derivative.append(approx_derivative)


central_derivative_array = np.array(central_derivative)
step_sizes_array = np.array(step_sizes)


errors = np.abs(central_derivative_array - exact_derivative)

# Plot 

plt.plot(step_sizes_array, errors, marker='o', label="Error in Central Derivative")
plt.xscale('log')  # Logarithmic scale for x-axis (step size)
plt.yscale('log')  # Logarithmic scale for y-axis (error)
plt.xlabel("Step size (h)")
plt.ylabel("Absolute Error")
plt.title("Error in Central Derivative Approximation vs Step Size")
plt.legend()
plt.show()
```

The optimal value of h=10^-6
